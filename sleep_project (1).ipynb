```python

def exercise_level(x):
    if x <= 30:
        return 'Low'
    elif x <= 60:
        return 'Moderate'
    else:
        return 'High'

def screen_time_level(x):
    if x <= 60:
        return 'Low'
    elif x <= 120:
        return 'Moderate'
    else:
        return 'High'


def caffeine_level(x):
    if x <= 100:
        return 'Low'
    elif x <= 200:
        return 'Moderate'
    else:
        return 'High'

def Sleep_Quality_level(x):
    if x <= 3:
        return 'Low'
    elif x <= 7:
        return 'Moderate'
    else:
        return 'High'

def age_group(x):
    if x <= 20:
        return '0-20'
    elif x <= 40:
        return '21-40'
    else:
        return '41-59'

df['Age_Group'] = [age_group(i) for i in df['Age']]
df['sleep_quality_level'] = [Sleep_Quality_level(i) for i in df['Sleep Quality']]
df['Exercise_Level'] = [exercise_level(i) for i in df['Exercise (mins/day)']]
df['Screen_Time_Level'] = [screen_time_level(i) for i in df['Screen Time Before Bed (mins)']]
df['Caffeine_Level'] =[caffeine_level(i) for i in df['Caffeine Intake (mg)']]
```

```python
plt.figure(figsize=(16,10))
plt.subplot(1,2,1)
sns.boxplot(x='Exercise_Level', y='Sleep Quality', data=df, palette='cool',hue='Gender')
plt.subplot(1,2,2)
sns.violinplot(x='Exercise_Level', y='Sleep Quality', data=df, palette='cool',hue='Gender')

plt.title('Sleep Quality by Exercise Level')
plt.xlabel('Exercise Level')
plt.ylabel('Sleep Quality')
plt.show()


```

```python

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

axes[0].hist(df['Age'], bins=10, edgecolor='black', alpha=0.7)
axes[0].set_title('Age Distribution (Histogram)')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Count')

age_counts = df['Age_Group'].value_counts()

axes[1].pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Age Group Distribution (Pie Chart)')


a = sns.barplot(x='Age_Group', y='Sleep Quality', data=df, alpha=0.8, ax=axes[2])
axes[2].set_title('Age Group vs Sleep Quality')
axes[2].set_xlabel('Age Group')
axes[2].set_ylabel('Average Sleep Score (1â€“10)')
for i in a.containers:
  a.bar_label(i)

plt.tight_layout()
plt.show()

```
